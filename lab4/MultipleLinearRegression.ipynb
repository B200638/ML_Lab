## Housing Case Study
Problem Statement:

Consider a real estate company that has a dataset containing the prices of properties in the Delhi region. It wishes to use the data to optimise the sale prices of the properties based on important factors such as area, bedrooms, parking, etc.

Essentially, the company wants â€”

To identify the variables affecting house prices, e.g. area, number of rooms, bathrooms, etc.

To create a linear model that quantitatively relates house prices with variables such as number of rooms, area, number of bathrooms, etc.

To know the accuracy of the model, i.e. how well these variables can predict house prices.

```python
#1 Import the dataset (Housing.csv) and look at the top five rows
#import pandas as pd
#df = pd.read_csv('Housing.csv')
import pandas as pd

# Load dataset
df = pd.read_csv("Housing.csv")

# Display first 5 rows
print(df.head())

```

```python
#2 Check the shape of the DataFrame (rows, columns)
print(df.shape)
```

```python
#3 Look at the data types of the columns
print(df.dtypes)
```

```python
#4 Check for missing  values if any, replace with appropriate values
print(df.isnull().sum())  # Check for missing values

# If there are missing values, replace them with the mean or median
df.fillna(df.mean(), inplace=True)  # Replace NaN with column mean
```

```python
#5 Prepare  X (independent variables) and y (dependent variable)
X = df.drop(columns=['price'])  # Excluding the target variable
y = df['price']  # Target variable
```

```python
#6 Visualise the relationship between the independent variables and the depenedent variable using scatterplots
import matplotlib.pyplot as plt

for col in X.columns:
    plt.scatter(X[col], y, alpha=0.5)
    plt.xlabel(col)
    plt.ylabel("Price")
    plt.title(f"Price vs {col}")
    plt.show()
```

```python
#7 Encoding categorical data in X
X = pd.get_dummies(X, drop_first=True)  # Convert categorical variables into dummy variables
```

```python
#8  Avoiding the Dummy Variable Trap
#drop_first=True in pd.get_dummies() avoid it
```

```python
#9 Apply feature scaling on numerical variables
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Training Data Shape:", X_train.shape, y_train.shape)
print("Testing Data Shape:", X_test.shape, y_test.shape)
```

```python
#10 Split Data into Training and Testing Sets (70%-Train 30%-Test) - X_train, y_train, X_test, y_test
from sklearn.linear_model import LinearRegression

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Print coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
#11 Fitting Multiple Linear Regression to the Training
y_train_pred = model.predict(X_train)
errors_train = y_train_pred - y_train
```

```python
#12 Predict on the train set and calculate the error = y_pred - y_train
plt.scatter(y_train_pred, errors_train, color='blue', alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--')  # Horizontal line at zero
plt.xlabel("Predicted Prices")
plt.ylabel("Residuals (Errors)")
plt.title("Residual Plot - Train Data")
plt.show()
```

```python
#13  Residual plot - plot y_pred in x axis and errors in y axis
y_test_pred = model.predict(X_test)
errors_test = y_test_pred - y_test
```

```python
#14  Predic on the test set
plt.scatter(y_test_pred, errors_test, color='green', alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--')  # Horizontal line at zero
plt.xlabel("Predicted Prices")
plt.ylabel("Residuals (Errors)")
plt.title("Residual Plot - Test Data")
plt.show()
```

```python
#15 Residual plot - plot y_pred in x axis and errors in y axis
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_test_pred)
r2 = r2_score(y_test, y_test_pred)

print(f"Mean Squared Error: {mse}")
print(f"R Squared Value: {r2}")
```

```python
#16 Print Mean Squared Error and R Squared Value
n = X_test.shape[0]  # Number of samples
k = X_test.shape[1]  # Number of independent variables

adj_r2 = 1 - ((1 - r2) * (n - 1) / (n - k - 1))

print(f"Adjusted R Squared Value: {adj_r2}")
```

```python
#17 Check  Adjusted R Squared Value(by selecting different no of input variables insted of all)

# Select only a subset of features (example: 'area', 'bedrooms', 'bathrooms')
selected_features = ['area', 'bedrooms', 'bathrooms']
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]

# Train model on selected features
model_selected = LinearRegression()
model_selected.fit(X_train_selected, y_train)

# Predict and calculate RÂ²
y_test_pred_selected = model_selected.predict(X_test_selected)
r2_selected = r2_score(y_test, y_test_pred_selected)

# Calculate Adjusted RÂ²
k_selected = len(selected_features)
adj_r2_selected = 1 - ((1 - r2_selected) * (n - 1) / (n - k_selected - 1))

print(f"R Squared Value (Selected Features): {r2_selected}")
print(f"Adjusted R Squared Value (Selected Features): {adj_r2_selected}")
```
