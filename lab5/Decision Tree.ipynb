## Handling Missing Values

```python
#Replace the msiing values in Embarked column by 'S'
train_data["Embarked"].fillna("S", inplace=True)
test_data["Embarked"].fillna("S", inplace=True)
```

```python
#Impute the missing values in Fare column with median value
train_data["Fare"].fillna(train_data["Fare"].median(), inplace=True)
test_data["Fare"].fillna(test_data["Fare"].median(), inplace=True)
```

```python
#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data

mean_age=train_data["Age"].mean()
std_age=train_data["Age"].std()
nan_count=train_data["Age"].isnull().sum()
train_data.loc[train_data["Age"].isnull(),"Age"]=np.random.normal(mean_age, std_age, nan_count)

test_mean_age=test_data["Age"].mean()
test_std_age=test_data["Age"].std()
test_nan_count=test_data["Age"].isnull().sum()
test_data.loc[test_data["Age"].isnull(), "Age"]=np.random.normal(test_mean_age, test_std_age, test_nan_count)
```

## Feature Engineering - Creating new features

```python
# Create and add a new feature FamilySize as a combination of SibSp and Parch
#FamilySize = Parch + SibSp + 1

train_data["FamilySize"]=train_data["SibSp"]+train_data["Parch"]+1
test_data["FamilySize"]=test_data["SibSp"]+test_data["Parch"]+1
```

```python
# Create new feature IsAlone from FamilySize
#If FamilySize == 1 then IsAlone = True else IsAlone = False
train_data["IsAlone"]=train_data["FamilySize"]==1
test_data["IsAlone"]=test_data["FamilySize"]==1
```

## Create a new attribute Has_cabin using the attribue Cabin. 
#if Cabin == NaN the Has_cabin = False else Has_cabin = True


```python
#Create a new attribute Has_cabin using the attribue Cabin.

#if Cabin == NaN the Has_cabin = False else Has_cabin = True
train_data['Has_cabin'] = train_data['Cabin'].notna()
test_data['Has_cabin'] = test_data['Cabin'].notna()
```

```python

#Drop the columns which may not be useful in prediction of survival
#Drop PassengerId, Name, Ticket, and Cabin

train_data.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"], inplace=True)
test_data.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"], inplace=True)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Load your dataset (Replace with actual dataset)
#df = pd.read_csv('your_dataset.csv')

# Example dataset (Iris dataset) for demonstration
from sklearn.datasets import load_iris
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target  # Target variable

# Step 1: Check and Clean Column Names
print("Column Names Before Cleaning:", df.columns)
df.columns = df.columns.str.strip()  # Remove extra spaces

# Step 2: Verify Target Column Exists
print("Updated Column Names:", df.columns)
target_col = "target"  # Replace with actual column name if different

# Step 3: Pairplot for Feature Visualization
sns.pairplot(df, hue=target_col)
plt.show()

# Step 4: Splitting Data into Train (90%) & Validation (10%) sets
X = df.drop(columns=[target_col])
y = df[target_col]
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.1, random_state=42)

# Step 5: Creating and Training Decision Tree Model
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Step 6: Visualizing Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(model, filled=True, feature_names=X.columns, class_names=[str(i) for i in df[target_col].unique()])
plt.show()

# Step 7: Model Predictions
y_pred = model.predict(X_validation)

# Step 8: Evaluation Metrics
print("\nClassification Report:\n", classification_report(y_validation, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_validation, y_pred))

# Accuracy Score
accuracy = accuracy_score(y_validation, y_pred)
print(f"\nAccuracy: {accuracy:.4f}")

# Extracting Precision, Recall, and F1-score
report = classification_report(y_validation, y_pred, output_dict=True)
precision = report['accuracy']
recall = report['macro avg']['recall']
f1_score = report['macro avg']['f1-score']

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1_score:.4f}")
```
