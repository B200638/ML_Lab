```python
# Week 7: K Nearest Neighbor (KNN)

# 1. Import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report

# 2. Import the train data
file_path = "Week7_test.csv"  # Ensure this file is in your working directory
df = pd.read_csv(file_path)

# 3. Check shape of the data
print("Shape of dataset:", df.shape)

# 4. Select only 10% of the data for experiment
df_sample = df.sample(frac=0.1, random_state=1)

# 5. Print shape of selected dataset
print("Shape of sampled dataset:", df_sample.shape)

# 6. Look at the top 5 rows
print(df_sample.head())

# 7. Find number of unique class labels
print("Number of unique class labels:", df_sample['label'].nunique())

# 8. Print number of examples for each class label
print("Examples per class:\n", df_sample['label'].value_counts())

# 9. Prepare X and Y
X = df_sample.drop('label', axis=1)  # Features
Y = df_sample['label']              # Labels

# 10. Visualize the image of the first 5 digits in X
for i in range(5):
    image = X.iloc[i].values.reshape(28, 28)  # Modify if your data is not 28x28
    plt.imshow(image, cmap='gray')
    plt.title(f"Label: {Y.iloc[i]}")
    plt.axis('off')
    plt.show()

# 11. Split the train dataset into training set (75%) and validation set (25%)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
print("Training set size:", X_train.shape)
print("Validation set size:", X_test.shape)

# 12. Fit KNN classifier to training data
# 13. Use default parameters: n_neighbors=5, metric='minkowski', p=2
knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn.fit(X_train, Y_train)

# 14. Predict the validation set results
Y_pred = knn.predict(X_test)

# 15. Print the confusion matrix
print("Confusion Matrix:\n", confusion_matrix(Y_test, Y_pred))

# 16. Print classification report and accuracy
print("\nClassification Report:\n", classification_report(Y_test, Y_pred))

# 17. Visualize the first 5 test digits and predicted labels
for i in range(5):
    image = X_test.iloc[i].values.reshape(28, 28)
    plt.imshow(image, cmap='gray')
    plt.title(f"Actual: {Y_test.iloc[i]}, Predicted: {Y_pred[i]}")
    plt.axis('off')
    plt.show()
```

```python
# Q1. Import all necessary libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Q2. Load the dataset 'Week7_test.csv'.
file_path = "Week7_test.csv"
df = pd.read_csv(file_path)
```

```python
# Q3. Print the shape of the dataset.
print("Shape of dataset:", df.shape)
```

```python
# Q4. Select 10% of the data using sample() method.
df_sample = df.sample(frac=0.1, random_state=1)
```

```python
# Q5. Print the shape of the sampled dataset.
print("Shape of sampled dataset:", df_sample.shape)
```

```python
# Q6. Display the first 5 rows of the sampled data.
df_sample.head()
```

```python
# Q7. Print the number of unique class labels in the data.
print("Number of unique class labels:", df_sample['label'].nunique())
```

```python
# Q8. Print the count of examples in each class.
print("Examples per class:\n", df_sample['label'].value_counts())
```

```python
# Q9. Separate data into X (features) and Y (label).
X = df_sample.drop('label', axis=1)
Y = df_sample['label']
```

```python
# Q10. Plot the first 5 digit images.
for i in range(5):
    plt.imshow(X.iloc[i].values.reshape(28, 28), cmap='gray')
    plt.title(f"Label: {Y.iloc[i]}")
    plt.axis('off')
    plt.show()
```

```python
# Q11. Split the data into 75% training and 25% test.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
print("Training set size:", X_train.shape)
print("Validation set size:", X_test.shape)
```

```python
# Q12. Train a KNN classifier with k=5.
knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn.fit(X_train, Y_train)
```

```python
# Q13. Predict test set values using the trained model.
Y_pred = knn.predict(X_test)
```

```python
# Q14. Print confusion matrix and classification report.
print("Confusion Matrix:\n", confusion_matrix(Y_test, Y_pred))
print("\nClassification Report:\n", classification_report(Y_test, Y_pred))
```

```python
# Q15. Plot the first 5 test images with actual and predicted labels.
for i in range(5):
    plt.imshow(X_test.iloc[i].values.reshape(28, 28), cmap='gray')
    plt.title(f"Actual: {Y_test.iloc[i]}, Predicted: {Y_pred[i]}")
    plt.axis('off')
    plt.show()
```
